```python
# ========================================
#  CELL 3: VERIFY API KEY
# ========================================

import os

api_key = os.environ.get("OPENAI_API_KEY")

if api_key:
    print("API key found")
    print("Starts with:", api_key[:12] + "********")
else:
    print("API key not found. Please check your ~/.zshrc file")
```

```python
# ========================================
#  CELL 4: CONNECT TO OPENAI
# ========================================

from openai import OpenAI
import os

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
print("OpenAI client connected")
```

```python
# ========================================
#  CELL 5: TEST OPENAI CONNECTION (PAID)
# ========================================

resp = client.responses.create(
    model="gpt-5-nano",
    input="Write a haiku about AI and wildfire resilience"
)

print("Model response:\n")
print(resp.output_text)
```
